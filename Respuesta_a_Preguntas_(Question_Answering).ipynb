```python
from transformers import pipeline

# Inicializamos el modelo para la tarea de Respuesta a Preguntas
# El modelo por defecto se descargará automáticamente.
qa_pipeline = pipeline("question-answering")

print("Modelo de Respuesta a Preguntas cargado.")
```

```python
contexto = """
La Torre Eiffel fue construida por Gustave Eiffel para la Exposición Universal de 1889 en París.
Su construcción comenzó en 1887 y se completó en poco más de dos años, el 31 de marzo de 1889.
Originalmente iba a ser desmantelada después de 20 años, pero se salvó gracias a su valor como antena de radio.
"""

pregunta = "¿Quién construyó la Torre Eiffel?"
```

```python
# Realizamos la predicción
resultado_qa = qa_pipeline(question=pregunta, context=contexto)

print("\nResultado:", resultado_qa)
```
